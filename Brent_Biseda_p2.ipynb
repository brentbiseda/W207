```python
def CalcFScore(model, dev):
    preds = model.predict(dev)
    fScore = metrics.f1_score(dev_labels, preds, average='weighted')
    return fScore

def P3():
    """
    Use Count Vectorizer to Report F1 Score for k nearest neighbors classifier
        Find Optimal Value of K
    Fit Multinomial Naive Bayes
        Find Optimal Alpha
    Fit Logistic Regression
        Find Optimal Regularization C using L2 Norm
    """
    
    #Create Count Vectorizers and Fit Data
    vectorizer = CountVectorizer()
    x = vectorizer.fit_transform(train_data).toarray()
    dev = vectorizer.transform(dev_data).toarray()
    
    #Train KNN
    model = KNeighborsClassifier()
    k = {'n_neighbors': [1, 3, 5, 7, 9, 15, 25]}
    model = GridSearchCV(model, k)
    model.fit(x, train_labels)
    bestN = model.best_params_['n_neighbors']
    print("Best K is: {}".format(bestN))
    
    #F Score
    fScore = CalcFScore(model, dev)
    print('The F score for K-NN: {:.4}\n'.format(fScore))
    
    #Multinomial
    model = MultinomialNB()
    alphas = {'alpha': [0.0001, 0.001, 0.01, 0.1, 0.5, 1.0, 2.0, 10.0]}
    model = GridSearchCV(model, alphas)
    model.fit(x, train_labels)
    bestAlpha = model.best_params_['alpha']
    print('Best alpha is: {}'.format(bestAlpha))

    #F Score
    fScore = CalcFScore(model, dev)
    print('The F score for Multi-nomial Naive Bayes: {:.4}\n'.format(fScore))
    
    #Logistic Regression
    cs = {'C': [0.001, 0.01, 0.1, 1.0, 2.0, 10.0]}
    model = LogisticRegression(penalty = 'l2', solver='lbfgs', multi_class='multinomial', max_iter=1000)
    model = GridSearchCV(model, cs)
    model.fit(x, train_labels)
    bestC = model.best_params_['C']
    print('Best C is: {}'.format(bestC))
        
    #F Score
    fScore = CalcFScore(model, dev)
    print('The F score for Logistic Regression: {:.4}\n'.format(fScore))
    
    #Print Sum of Squares of C Values
    for i in cs['C']:
        model = LogisticRegression(penalty = 'l2', C=i, solver='lbfgs', multi_class='multinomial', max_iter=1000)
        model.fit(x, train_labels)
        weights = [sum(y) for y in model.coef_**2]
        print('For C value of {}, the sums of weight vectors squared are: \n{} '.format(i, weights)) 
    
P3()
```

(4) Train a logistic regression model. Find the 5 features with the largest weights for each label -- 20 features in total. Create a table with 20 rows and 4 columns that shows the weight for each of these features for each of the labels. Create the table again with bigram features. Any surprising features in this table?

[5 pts]

```python
import pandas as pd

def getKeysByValue(dictOfElements, valueToFind):
    listOfItems = dictOfElements.items()
    for item in listOfItems:
        if item[1] == valueToFind:
            listOfKeys = item[0]
    return  listOfKeys

def TabulateLogistic(n=1):
    vectorizer = CountVectorizer(ngram_range = (n, n))
    x = vectorizer.fit_transform(train_data).toarray()
    dev = vectorizer.transform(dev_data).toarray()

    #Logistic Regression
    model = LogisticRegression(penalty = 'l2', C=0.1, solver='lbfgs', multi_class='multinomial', max_iter=1000)
    model.fit(x, train_labels)
    featureWeights = model.coef_

    #Determine largest coefficient indices for each category
    largeWeights = []
    for col in featureWeights:
        indices = [x for x in np.argsort(col)[-5:]]
        largeWeights += indices

    rowNames = [getKeysByValue(vectorizer.vocabulary_, y) for y in largeWeights]

    #create array that will be turned into table
    arrayData = []
    for index in largeWeights:
        #Get Coefficients from Row
        row = [featureWeights[0][index], featureWeights[1][index], featureWeights[2][index], featureWeights[3][index]]
        arrayData.append(row)
        #Create Pandas Dataframe
    return pd.DataFrame(arrayData, index = rowNames, columns=newsgroups_train.target_names)

def P4():
    """
    Train a logistic regression model. 
    Find the 5 features with the largest weights for each label -- 20 features in total. 
    Create a table with 20 rows and 4 columns that shows the weight for each of these features for each of the labels. 
    Create the table again with bigram features. Any surprising features in this table?
    """
    print('For Unigram:')
    df1 = TabulateLogistic(n=1)
    print(df1)

    print('\nFor Bigram:')
    df2 = TabulateLogistic(n=2)
    print(df2)
    
P4()
```

The preprocessing function runs on the raw text, before it is split into words by the tokenizer. Your preprocessor should try to normalize the input in various ways to improve generalization. For example, try lowercasing everything, replacing sequences of numbers with a single token, removing various other non-letter characters, and shortening long words. If you're not already familiar with regular expressions for manipulating strings, see https://docs.python.org/2/library/re.html, and re.sub() in particular.
```python
import string

def empty_preprocessor(s):
    return s

def better_preprocessor(s):
    #Lowercase everything
    s = s.lower()
    #Replace all numbers
    s = re.sub('(\d)+', 'NUM', s)
    #Remove all punctuation
    table = str.maketrans({key: None for key in string.punctuation})
    s = s.translate(table)
    # Make all white space a single space
    s = re.sub("\s+", " ", s)
    # Remove any trailing or leading white space
    s = s.strip(" ")

    return s

def CalcFScore(model, dev):
    preds = model.predict(dev)
    fScore = metrics.f1_score(dev_labels, preds, average='weighted')
    return fScore

def P5():
    #No Preprocessor
    vectorizer = CountVectorizer(preprocessor=empty_preprocessor)
    x = vectorizer.fit_transform(train_data).toarray()
    dev = vectorizer.transform(dev_data).toarray()
    model = LogisticRegression(penalty = 'l2', solver='lbfgs', multi_class='multinomial', max_iter=1000)
    model.fit(x, train_labels)
    fScore = CalcFScore(model, dev)
    wordLengthEmpty = len(vectorizer.vocabulary_)
    print("F Score with no Preprocessor: {:.4}".format(fScore))
    
    #Improved Preprocessor
    vectorizer = CountVectorizer(preprocessor=better_preprocessor)
    x = vectorizer.fit_transform(train_data).toarray()
    dev = vectorizer.transform(dev_data).toarray()
    model = LogisticRegression(penalty = 'l2', solver='lbfgs', multi_class='multinomial', max_iter=1000)
    model.fit(x, train_labels)
    fScore = CalcFScore(model, dev)
    wordLengthImproved = len(vectorizer.vocabulary_)
    print("F Score with Improved Preprocessor: {:.4}".format(fScore))
    
    deltaWords = wordLengthEmpty - wordLengthImproved
    print("The Vocabulary was Reduced by: {}".format(deltaWords))
    
P5()
```

(6) The idea of regularization is to avoid learning very large weights (which are likely to fit the training data, but not generalize well) by adding a penalty to the total size of the learned weights. That is, logistic regression seeks the set of weights that minimizes errors in the training data AND has a small size. The default regularization, L2, computes this size as the sum of the squared weights (see P3, above). L1 regularization computes this size as the sum of the absolute values of the weights. The result is that whereas L2 regularization makes all the weights relatively small, L1 regularization drives lots of the weights to 0, effectively removing unimportant features.

```python
def getKeysByValue(dictOfElements, valueToFind):
    listOfItems = dictOfElements.items()
    for item in listOfItems:
        if item[1] == valueToFind:
            listOfKeys = item[0]
    return  listOfKeys

def Problem6ModelFit(c):
    print("For a value of c: {}".format(c))
    
    #Create Vectorizer
    vectorizer = CountVectorizer()
    x = vectorizer.fit_transform(train_data).toarray()
    dev = vectorizer.transform(dev_data).toarray()
    
    #Create Logistic Model with L1 Norm
    model1 = LogisticRegression(penalty = 'l1', solver='saga', multi_class='multinomial', max_iter=1000, C = c, tol=0.01)
    model1.fit(x, train_labels)
    
    nonZeroL1 = np.count_nonzero(model1.coef_)
    print('Non-zero weights using L1: {}'.format(nonZeroL1))
    
    #Identify non-zero
    uniqueL1 = np.nonzero(model1.coef_)[1].tolist()
    #Keep Only unique values
    uniqueL1 = np.unique(uniqueL1)
    
    #Create Logistic Model with L2 Norm
    model2 = LogisticRegression(penalty = 'l2', solver='saga', multi_class='multinomial', max_iter=1000, C = c, tol=0.01)
    model2.fit(x, train_labels)
    
    nonZeroL2 = np.count_nonzero(model2.coef_)
    print('Non-zero weights using L2: {}'.format(nonZeroL2))
    
    #Create New Vocab from L1 Words
    newVocab = [getKeysByValue(vectorizer.vocabulary_, i) for i in uniqueL1]
    
    #Create New Vectorizer That only Uses L1 Words
    vectorizerNew = CountVectorizer(vocabulary=newVocab)
    x = vectorizerNew.fit_transform(train_data).toarray()
    dev = vectorizerNew.transform(dev_data).toarray()
    
    model3 = LogisticRegression(penalty = 'l2', solver='saga', multi_class='multinomial', max_iter=1000, C = c, tol=0.01)
    model3.fit(x, train_labels)
    preds = model3.predict(dev)
    
    vocabsize = len(newVocab)
    accuracy = metrics.accuracy_score(dev_labels, preds) * 100
    print("Pruning Features from L1 then Fitting with L2: ")
    print("Vocab Size of: {}".format(vocabsize))
    print("Accuracy of: {:.5}%".format(accuracy))
    print("")
    
    return (vocabsize, accuracy)

def P6():
    """
    Use L1 & L2 Regularizers to experiment with number of non-zero features
    """
    
    # Keep this random seed here to make comparison easier.
    np.random.seed(0)
    
    cs = [.01,.1,1.0,10,25,50,75,100]
    accuracies = []
    vocabsizes = []
    for c in cs:
        vocab, accuracy = Problem6ModelFit(c)
        vocabsizes.append(vocab)
        accuracies.append(accuracy)
        
    plt.plot(vocabsizes, accuracies, 'ro')
    plt.xlabel('Vocab Size')
    plt.ylabel('Accuracy')
    plt.title('Effect of Vocab Size on Accuracy')

P6()
```
